```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

data = {
    'Student': ['Amara', 'Ben', 'Chen', 'Divya', 'Ethan', 'Fatima', 'George',
                'Hana', 'Ivan', 'Jade', 'Kwame', 'Lily'],
    'Test_Score': [62, 65, 68, 70, 71, 72, 74, 75, 12, 78, 80, 130]
}
df = pd.DataFrame(data)
df
```

```python
df['Test_Score'].describe()
```

```python
df.sort_values('Test_Score')
```

```python
Q1 = df['Test_Score'].quantile(0.25)
Q3 = df['Test_Score'].quantile(0.75)
IQR = Q3 - Q1
print(Q1, Q3, IQR)
```

```python
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['Test_Score'] < lower) | (df['Test_Score'] > upper)]
outliers
```

```python
plt.boxplot(df['Test_Score'], vert=False)
plt.title('Boxplot of Test_Score')
plt.xlabel('Test_Score')
plt.show()
```

```python
mean_score = df['Test_Score'].mean()
std_score = df['Test_Score'].std()
print(mean_score, std_score)
```

```python
df['Z_Score'] = stats.zscore(df['Test_Score'])
df[(df['Z_Score'] > 3) | (df['Z_Score'] < -3)]
```

```python
df['Test_Score'].plot(kind='hist', bins=6, edgecolor='black')
plt.title('Distribution of Test_Score')
plt.xlabel('Test_Score')
plt.show()
```

```python
df_clean = df.drop(outliers.index)
df_clean
```

```python
df['Test_Score_Capped'] = df['Test_Score'].clip(lower=lower, upper=upper)
df[['Student', 'Test_Score', 'Test_Score_Capped']]
```
